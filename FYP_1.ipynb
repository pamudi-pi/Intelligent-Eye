```python
# Install the ultralytics package
!pip install ultralytics

# Import necessary libraries
from ultralytics import YOLO
from google.colab import files
import cv2
import os

# Upload the video file
uploaded = files.upload()  # Allow user to upload a video file

# Get the uploaded file name
video_file = list(uploaded.keys())[0]
print(f"Uploaded file: {video_file}")

# Load the YOLO model (you can replace 'yolov8n.pt' with a custom model)
model = YOLO('yolov8n.pt')  # YOLOv8 nano model

# Run the model on the uploaded video file
results = model(
    source=video_file,   # Path to the uploaded video file
    show=True,           # Display the processed frames (useful for debugging)
    conf=0.4,            # Confidence threshold for detections
    save=True            # Save the processed video
)

# Find the processed video in the 'runs/detect/predict' directory
output_dir = 'runs/detect/predict'
output_video = [f for f in os.listdir(output_dir) if f.endswith('.mp4')][0]
output_path = os.path.join(output_dir, output_video)

print(f"Processed video saved as: {output_path}")

# Allow download of the processed video
files.download(output_path)
```

```python
# Install the ultralytics package
!pip install ultralytics

# Import necessary libraries
from ultralytics import YOLO
from google.colab import files
import cv2
import os
from IPython.display import HTML

# Upload the video file
uploaded = files.upload()  # Allow user to upload a video file

# Get the uploaded file name
video_file = list(uploaded.keys())[0]
print(f"Uploaded file: {video_file}")

# Load the YOLO model (you can replace 'yolov8n.pt' with a custom model)
model = YOLO('yolov8n.pt')  # YOLOv8 nano model

# Run the model on the uploaded video file
results = model(
    source=video_file,   # Path to the uploaded video file
    show=True,           # Display the processed frames (useful for debugging)
    conf=0.4,            # Confidence threshold for detections
    save=True            # Save the processed video
)

# Find the processed video in the 'runs/detect/predict' directory
output_dir = 'runs/detect/predict'

if os.path.exists(output_dir):
    # Find .mp4 files in the directory
    output_files = [f for f in os.listdir(output_dir) if f.endswith('.mp4')]

    if output_files:
        output_video = output_files[0]  # Take the first .mp4 file
        output_path = os.path.join(output_dir, output_video)

        print(f"Processed video saved as: {output_path}")

        # Provide a clickable link for downloading
        print("Click the link below to download the processed video:")
        display(HTML(f'<a href="{output_path}" download>Download Processed Video</a>'))
    else:
        print("No .mp4 files found in the output directory.")
else:
    print(f"Output directory '{output_dir}' does not exist.")
```
